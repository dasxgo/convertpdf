## Extracting tables from a PDF

```python
import pdfplumber
import pandas as pd
from PIL import Image
```

```python
pdf = pdfplumber.open("ex900942130.pdf")

```

## Load page_0

```python
p0 = pdf.pages[0]
```

## Visualice the actual content on page_0

```python
Image.open("page_0.png")
```

## Extract the table and load as a panda df

```python
table = p0.extract_table()

df =pd.DataFrame(table[1:], columns=table[0])
for column in ["CLIENTE", "NIT"]:
    df[column] = df[column].str.replace("","")
```
